```python
import requests
import pandas as pd
import bs4
req=requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup=bs4.BeautifulSoup(req.text)
a=[]
b=[]
c=[]
x=soup.find("table").find_all_next('tr')
for i in x[1:-4]:
    j=i.find_all('td')
    a.append(j[0].text.strip())
    b.append(j[1].text.strip())
    c.append(j[2].text.strip())
df=pd.DataFrame(data={"Postal Code":a,"Borough":b,"Neighbourhood":c})
df=df[df['Borough']!="Not assigned"]
df.head(15)
```
